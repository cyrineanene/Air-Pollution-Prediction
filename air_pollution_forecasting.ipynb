In this dataset, according to the UCI documentation 'Missing values are tagged with -200 value.'

```python
#counting the number of true missing data
df1.isin([-200]).sum(axis=0)
```

Handling those missing values

```python
#changing the -200 to NaN then to the mean value
df1 = df1.replace(to_replace=-200, value= np.nan)
```

```python
print(df1.isin([-200]).sum(axis=0))
print(df1.isnull().sum())
```

```python
df1 = df1.fillna(df1.mean)
```

```python
df1.isnull().sum()
```

#### Feature Engineering

1. Adding the `AQI Value` Column as it will serve later as our Y Target Column.
2. Changing the format of the Date and Time columns as it will serve later as our DS Column.

```python
#Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(df1['Date'], dayfirst=True)
date_info
```

```python
#Converting the time from HH.MM.SS to HH:MM:SS
time_info = df1['Time'].apply(lambda x: x.replace('.', ':'))
time_info
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining the 2 series (date_info and time_info) to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis = 1)
print(date_time.head(5))
print(date_time.shape)
```

```python
#combining the two columns into a single one, the column should be named ds as Prophet won't understand any other names
date_time['ds'] = date_time['Date'].astype(str) + ' '+ date_time['Time'].astype(str)
date_time.head(5)
```

```python
#Converting DS from Object Datatype to DateTime Format
data = pd.DataFrame()
data['ds'] = pd.to_datetime(date_time['ds'])
data.info()
```

#### Forecasting Phase with FB Prophet

**Time Series Forecasting**: The process of predicting the future values of the data by analyzing the previous trends and patterns hidden in the data.
Time series forecasting can be done using various forecasting techniques, in our case we will be using `FB Prophet`.

We will continue working on the data DataFrame and adding to it our desired Y Column.

```python
df1['RH'] = pd.to_numeric(df1['RH'], errors='coerce')
data['y'] = df1['RH']
data.info()
```

```python
from prophet import Prophet
```

```python
#Training the model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
#'yhat' is the middle or the predicted value, 'yhat_lower'is the min, 'yhat_upper' is the max
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#Figure based on 'yhat', 'yhat_lower' and 'yhat_upper' columns
fig1 = model.plot(forecast)
```

```python
#Figure based on the predictions for Week/Day/Time and the first plot is for the trend
fig2 = model.plot_components(forecast)
```

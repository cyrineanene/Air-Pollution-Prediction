#### Feature Engineering

1. Adding the `AQI Value` Column as it will serve later as our Y Target Column.
2. Changing the format of the Date and Time columns as it will serve later as our DS Column.

```python
#Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(df1['Date'], dayfirst=True)
date_info
```

```python
#Converting the time from HH.MM.SS to HH:MM:SS
time_info = df1['Time'].apply(lambda x: x.replace('.', ':'))
time_info
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining the 2 series (date_info and time_info) to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis = 1)
print(date_time.head(5))
print(date_time.shape)
```

```python
#combining the two columns into a single one, the column should be named ds as Prophet won't understand any other names
date_time['ds'] = date_time['Date'].astype(str) + ' '+ date_time['Time'].astype(str)
date_time.head(5)
```

```python
#Converting DS from Object Datatype to DateTime Format
data = pd.DataFrame()
data['ds'] = pd.to_datetime(date_time['ds'])
data.info()
```

```python
#We will be working on the CO and NO2 Columns to calculate the overall AQI Value

# Defining the breakpoints for CO according to EPA standards
breakpoints_co = [
    (0.0, 4.4, 0, 50),
    (4.5, 9.4, 51, 100),
    (9.5, 12.4, 101, 150),
    (12.5, 15.4, 151, 200),
    (15.5, 30.4, 201, 300),
    (30.5, 40.4, 301, 400),
    (40.5, 50.4, 401, 500)
]

# Defining the breakpoints for NO2 in according to EPA standards
breakpoints_no2 = [
    (0, 53, 0, 50),
    (54, 100, 51, 100),
    (101, 360, 101, 150),
    (361, 649, 151, 200),
    (650, 1249, 201, 300),
    (1250, 1649, 301, 400),
    (1650, 2049, 401, 500)
]

def calculate_aqi(concentration, breakpoints):
    """Calculate AQI for a given concentration based on breakpoints."""
    for c_low, c_high, i_low, i_high in breakpoints:
        if c_low <= concentration <= c_high:
            aqi = ((i_high - i_low) / (c_high - c_low)) * (concentration - c_low) + i_low
            return round(aqi)
    #if the concentration is outside the defined ranges
    return None  
df1['CO(GT)'] = pd.to_numeric(df1['CO(GT)'], errors='coerce')
df1['NO2(GT)'] = pd.to_numeric(df1['NO2(GT)'], errors='coerce')
df1['AQI_CO'] = df1['CO(GT)'].apply(lambda x: calculate_aqi(x, breakpoints_co))
df1['AQI_NO2'] = df1['NO2(GT)'].apply(lambda x: calculate_aqi(x, breakpoints_no2))
df1['Overall_AQI'] = df1[['AQI_CO', 'AQI_NO2']].max(axis=1)
```

#### Forecasting Phase with FB Prophet

**Time Series Forecasting**: The process of predicting the future values of the data by analyzing the previous trends and patterns hidden in the data.
Time series forecasting can be done using various forecasting techniques, in our case we will be using `FB Prophet`.

We will continue working on the data DataFrame and adding to it our desired Y Column.

```python
data['y'] = df1['Overall_AQI']
data.info()
```

```python
from prophet import Prophet
```

```python
#Training the model
model = Prophet(
    changepoint_prior_scale=0.1, 
    seasonality_prior_scale=15.0, 
    holidays_prior_scale=10.0
)
model.fit(data)
```

```python
from sklearn.model_selection import ParameterGrid

param_grid = {
    'changepoint_prior_scale': [0.05, 0.1, 0.2],
    'seasonality_prior_scale': [5.0, 10.0, 15.0],
    'holidays_prior_scale': [5.0, 10.0, 20.0]
}

grid = ParameterGrid(param_grid)
for params in grid:
    model = Prophet(**params)
    model.fit(data)
    # Evaluate model here
    from prophet.diagnostics import cross_validation
    from prophet.diagnostics import performance_metrics

    df_cv = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    df_p = performance_metrics(df_cv)
    print(df_p)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
#'yhat' is the middle or the predicted value, 'yhat_lower'is the min, 'yhat_upper' is the max
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#Figure based on 'yhat', 'yhat_lower' and 'yhat_upper' columns
fig1 = model.plot(forecast)
```

The forecast shows a reasonable fit with the actual data (black dots), though there is variability within the prediction intervals (shaded blue regions). The model captures the seasonal patterns (weekly and daily) but may slightly overfit the daily fluctuations, as shown by the dense oscillations.

```python
#Figure based on the predictions for Week/Day/Time and the first plot is for the trend
fig2 = model.plot_components(forecast)
```

1. The first plot shows the overall trend of the variable over time: The trend shows a clear progression over time. There is a dip observed around late 2004 but a recovery afterward, stabilizing in early 2005. 
=> This indicates that the model has captured an overall underlying trend in your data effectively.

2. The second plot shows the variable's fluctuation across the days of the week: The weekly pattern demonstrates periodic behavior, with activity peaking on Friday and declining sharply on Sunday. The AQI Value varies systematically during the week.

3. This third chart represents the variation of the variable across different hours of the day: The daily pattern shows variations throughout the day, with the lowest values early in the morning and peaks around the evening. This pattern may correspond to human activity, such as increased emissions during commuting hours.
